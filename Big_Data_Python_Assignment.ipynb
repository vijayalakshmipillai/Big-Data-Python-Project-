```python

import numpy as np
import pandas as pd
from sklearn import linear_model, preprocessing
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn import feature_selection, metrics
```

```python
def classify_for_threshold(clf, testX, testY, t):
    prob_df = pd.DataFrame(clf.predict_proba(testX)[:, 1])
    prob_df['predict'] = np.where(prob_df[0]>=t, 1, 0)
    prob_df['actual'] = testY
    return pd.crosstab(prob_df['actual'], prob_df['predict'])
```

```python
(temp2['award'].value_counts())/temp2['award'].size
```

```python
temp2['award'] = (temp2['award']=='Best Director').astype(int)
```

```python
cat_vars = ['race_ethnicity', 'Country']
temp2 = pd.get_dummies(temp2, columns=cat_vars)
model = linear_model.LogisticRegression()
X0 = temp2.loc[:, temp2.columns != 'award']
Y0 = temp2['award']
selector = feature_selection.RFE(model, n_features_to_select=5, step=1)
selector = selector.fit(X0, Y0)
r_features = X0.loc[:, selector.support_]
print("R features are:\n{}".format(','.join(list(r_features))))
```

```python
X = r_features
Y = temp2['award']
trainX, testX, trainY, testY = train_test_split(X, Y, test_size=0.3, random_state=0)
clf = linear_model.LogisticRegression()
clf.fit(trainX, trainY)
print(' {}'.format(clf.intercept_))
```

```python
print(' {}'.format(clf.coef_))
```

```python
predicted = clf.predict(testX)
print("Mean hits: {}".format(np.mean(predicted==testY)))
```

```python
scores = cross_val_score(linear_model.LogisticRegression(), X, Y, scoring='accuracy', cv=8)
scores
```

```python
print("Mean scores: {}".format(scores.mean()))
```

```python
xtab = classify_for_threshold(clf, testX, testY, 0.2)
print("Threshold {}:\n{}\n".format(0.2, xtab))
```

```python
temp4_norm = (temp4 - temp4.min()) / (temp4.max() - temp4.min())
temp4_norm.head(2)
```

```python
model = KMeans(n_clusters=6)
model.fit(temp4_norm)
print('J-score = ', model.inertia_)
print(model.labels_)
```

```python
labels = model.labels_
md = pd.Series(labels)
temp4_norm['clust'] = md
temp4_norm.head(5)
```

```python
centroids = model.cluster_centers_
print ('centroids', centroids)
```

```python
plt.hist(temp4_norm['clust'])
plt.title('Histogram of Clusters')
plt.xlabel('Cluster')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```
